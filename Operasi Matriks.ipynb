```python
A = [[5, 0, 4],
     [2, 6, 2],
     [1, 4, 1],]

B =  [[1, 0, 3],
      [4, 2, 1],
      [2, 3, 1],]

C = []
```

```python
#penjumlahan
for x in range(len(A)):
  for y in range(len(A[0])):
    print(A[x][y] + B[x][y], end=' ')
  print()
```

```python
#pengurangan
for x in range(len(A)):
  for y in range(len(A[0])):
    print(A[x][y] - B[x][y], end=' ')
  print()
```

```python
A = [
    [2, 3],
    [0, 9],
    [7, 1]
]

B = [
    [10, 1],
    [-6, 4]
]

C = []

# perkalian matriks
for x in range(len(A)):          # baris A
    row = []
    for y in range(len(B[0])):   # kolom B
        total = 0
        for z in range(len(B)):  # kolom A / baris B
            total += A[x][z] * B[z][y]
        row.append(total)
    C.append(row)

# tampilkan hasil
for x in range(len(C)):
    for y in range(len(C[0])):
        print(C[x][y], end=' ')
    print()

```

```python
#perkalian dengan skalar

k = 2
A = [[2, 5, 4],
     [1, 7, 3]]

C = []

for i in range(len(A)):
    row= []
    for j in range(len(A[i])):
        row.append(k * A[i][j])
    C.append(row)

print("Hasil: ")
for row in C:
    print(row)
```

```python
#transpose (3 x 3)
A = [[5, 0, 4],
     [2, 6, 2],
     [1, 4, 1]]

Bt = []

for row in zip(*A):
    Bt.append(list(row))

for x in range(len(Bt)):
    for y in range(len(Bt[0])):

     print(Bt[x][y], end=' ')
    print()
```

```python
#trace
def trace(matrix):
    return sum(matrix[i][i] for i in range(len(matrix)))

A = [
    [4, 9],
    [1, 3]
]

B = [
    [1, 0, 3],
    [4, 2, 1],
    [2, 3, 1]
]

print("trace(A) =", trace(A))
print("trace(B) =", trace(B))
```
